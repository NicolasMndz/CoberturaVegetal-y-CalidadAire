```python
import pandas as pd
```

```python
df = pd.read_excel('correlacion.xlsx')
display(df.head())
```

```python
correlation = df['Cobertura vegetal'].corr(df['Contaminante'])
print(f"The correlation between 'Cobertura vegetal' and 'Contaminante' is: {correlation}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8)) # Increased figure size for better readability with multiple colors
sns.scatterplot(data=df, x='Cobertura vegetal', y='Contaminante', hue='Estación')
plt.title('Scatter Plot de Cobertura vegetal vs Contaminante por Estación')
plt.xlabel('Cobertura vegetal')
plt.ylabel('Contaminante')
plt.grid(True)
plt.show()
```
